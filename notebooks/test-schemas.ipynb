## Тест Resume

```python
resume_dict = {
    "resume_id": 123,
    "job_title": "Python разработчик",
    "grade": "middle",
    "skills_res": "Python, Django, FastAPI, PostgreSQL",
    "about_me": "Опыт бэкенда 3 года",
    "exp_text": "Разрабатывал API, микросервисы, Docker",
    "exp_count": 3,
    "location": "Москва",
    "salary_val": 200000,
    "edu_uni": "МГУ",
    "edu_year": 2020,
}

resume = Resume.from_dict(resume_dict)

print("📄 RESUME")
print(f"ID: {resume.resume_id}")
print(f"Должность: {resume.job_title}")
print(f"Грейд: {resume.grade}")
print(f"Навыки: {resume.skills_res}")
print()

print("🔍 Текст для поиска:")
print(resume.to_search_text())
print()

print("📤 В словарь:")
print(resume.to_dict())
print()

assert resume.resume_id == 123
assert "Python" in resume.to_search_text()

empty_resume = Resume.from_dict({"resume_id": 999})
print(f"Пустое резюме: {empty_resume.to_search_text()}...")

none_resume = Resume.from_dict({"resume_id": 998, "edu_uni": None, "edu_year": None})
print(f"None поля: edu_uni={none_resume.edu_uni}, edu_year={none_resume.edu_year}")

print("✅ Resume тесты пройдены\n")
```

## Тест MatchResult

```python
mr1 = MatchResult(
    resume_id=123,
    matches=[
        VacancyMatch(vacancy_id=1, job_title="Python Dev", target_role="Backend", score=0.85, salary="100000 - 150000"),
        VacancyMatch(vacancy_id=2, job_title="Java Dev", target_role="Backend", score=0.72, salary="от 200000"),
        VacancyMatch(vacancy_id=3, job_title="Go Dev", target_role="Backend", score=0.65, salary="до 180000"),
    ],
    status="success",
)
assert mr1.resume_id == 123, "❌ resume_id"
assert len(mr1.matches) == 3, "❌ количество матчей"
assert mr1.status == "success", "❌ status"
print("✅ Тест 4: MatchResult создание")
```

```python
mr_data = mr1.to_dict()
assert "resume_id" in mr_data, "❌ Нет resume_id"
assert "matches" in mr_data, "❌ Нет matches"
assert "status" in mr_data, "❌ Нет status"
assert len(mr_data["matches"]) == 3, "❌ количество матчей в dict"
assert mr_data["matches"][0]["score"] == 0.85, "❌ score первого матча"
print("✅ Тест 5: MatchResult.to_dict()")
```

```python
mr_dict = {
    "resume_id": 456,
    "matches": [
        {"vacancy_id": 10, "job_title": "DevOps", "target_role": "Infrastructure", "score": 0.91, "salary": "200000 - 300000"},
        {"vacancy_id": 11, "job_title": "SRE", "target_role": "Infrastructure", "score": 0.78, "salary": "от 250000"},
    ],
    "status": "success",
}
mr2 = MatchResult.from_dict(mr_dict)
assert mr2.resume_id == 456, "❌ resume_id"
assert len(mr2.matches) == 2, "❌ количество матчей"
assert mr2.matches[0].vacancy_id == 10, "❌ vacancy_id первого матча"
assert mr2.matches[0].score == 0.91, "❌ score первого матча"
assert isinstance(mr2.matches[0], VacancyMatch), "❌ Тип матча не VacancyMatch"
print("✅ Тест 6: MatchResult.from_dict()")
```

```python
mr_empty = MatchResult(
    resume_id=999,
    matches=[],
    status="no_matches",
)
assert mr_empty.resume_id == 999, "❌ resume_id"
assert len(mr_empty.matches) == 0, "❌ матчи не пустые"
assert mr_empty.status == "no_matches", "❌ status"
print("✅ Тест 7: Пустой результат (no_matches)")
```

```python
mr_original = MatchResult(
    resume_id=789,
    matches=[
        VacancyMatch(vacancy_id=100, job_title="ML Engineer", target_role="Data Science", score=0.88, salary="300000 - 500000"),
    ],
    status="success",
)
json_str = json.dumps(mr_original.to_dict(), ensure_ascii=False)
mr_restored = MatchResult.from_dict(json.loads(json_str))
assert mr_restored.resume_id == mr_original.resume_id, "❌ resume_id после JSON"
assert len(mr_restored.matches) == 1, "❌ количество матчей после JSON"
assert mr_restored.matches[0].score == 0.88, "❌ score после JSON"
print("✅ Тест 8: Сериализация в JSON и обратно")
```

```python
mr_unsorted = MatchResult(
    resume_id=111,
    matches=[
        VacancyMatch(vacancy_id=1, job_title="Low", target_role="", score=0.50, salary=""),
        VacancyMatch(vacancy_id=2, job_title="High", target_role="", score=0.90, salary=""),
        VacancyMatch(vacancy_id=3, job_title="Medium", target_role="", score=0.70, salary=""),
    ],
    status="success",
)
# Проверяем что матчи есть (сортировка делается в Matcher, не в схеме)
scores = [m.score for m in mr_unsorted.matches]
assert len(scores) == 3, "❌ количество матчей"
assert 0.90 in scores and 0.70 in scores and 0.50 in scores, "❌ scores не совпадают"
print("✅ Тест 10: Матчи содержат score (сортировка в Matcher)")
```
